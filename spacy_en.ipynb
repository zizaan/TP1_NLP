```python
text = 'Perhaps one of the most important developments made by Arabic mathematics that began at this time with Al-Khwarizmi s work was the beginnings of algebra, and it is important to understand how important this new idea was, as it was a revolutionary step away from the Greek concept of mathematics which is essentially geometry. Algebra was a unifying theory that allowed rational numbers, irrational numbers, geometric expressions and others to be treated as algebraic objects, and gave mathematics as a whole a new path to development in a much broader concept than what had existed before, and provided a means for development in this subject The future and another important aspect of introducing the ideas of algebra is that it allowed the application of mathematics on itself in a way that had not happened before, and provided a means for development in this subject in the future'
```

```python
doc = nlp(text)

token_list = []
for token in doc:
           token_list.append(token.text)
print (token_list)
```

```python
filtered_sent=[]

for word in doc:
           if word.is_stop==False and word.text.isalpha()==True:
                      filtered_sent.append(word)
print(filtered_sent)

```

```python
post_tag=[]

for post in doc:
           post_tag.append([post.text,post.pos_])
print (post_tag)
```
